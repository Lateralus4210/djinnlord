```python
class EventDetermine:
    
    def random(event=None): #previously roll_for_encounter

        random_event_number = random.randint(1, 100 * 100) #(100% * Multiplier 100)
        #event_occurs = random_event_number <= probability
        print("Random event #" + str(random_event_number))
    
        EventDetermine.go(random_event_number)
    
    
    def go(event_number):
    
        multiplier = 100
    
        print("Determining event type for event " + str(event_number))
    
        enemy_encounter_chance = 80 * multiplier  # enemy
        item_encounter_chance = 35 * multiplier # Item
        rare_item_encounter_chance = 10 * multiplier # Rare Item
        shop_encounter_chance = 5 * multiplier # Shop
        zen_encounter_chance = 1 * multiplier # Zen
        
        if event_number <= enemy_encounter_chance:
            print("Event 1 occurs: Enemy")
            CharacterProfile.in_battle = True
    
            CommandsFrame.entry_listbox.insert(0, "Battle begin!")
            Travel.count_enemy_encounter += 1
            EventEnemy.determine(event_number)
    
        elif event_number <= item_encounter_chance:
            print("Event 2 occurs: Item")
            Travel.count_item_encounter += 1
            debug_listbox.insert(0, "Item event.")
            # EventItem.determine(event_number)

        elif event_number <= rare_item_encounter_chance:
            print("Event 3 occurs: Rare Item")
            Travel.count_rare_item_encounter += 1
            debug_listbox.insert(0, "Rare item event.")
            # EventItem.determine(event_number)
    
        elif event_number <= shop_encounter_chance:
            print("Event 4 occurs: Shop")
            Travel.count_shop_encounter += 1
            debug_listbox.insert(0, "Shop event.")
            # EventItem.determine(event_number)
            
        elif event_number <= zen_encounter_chance:
            print("Event 5 occurs: Zen")
            Travel.count_zen_encounter += 1
            debug_listbox.insert(0, "Zen event.")
            # EventItem.determine(event_number)
    
        else:
            print("No event occurs")
            Travel.count_none_event += 1
            debug_listbox.insert(0, "No event.")
            # EventItem.determine(event_number)
    
        print("\n")
    
    #Item Encounter: 8101 to 11500
    #Rare Item Encounter: 11601 to 12500
    #Shop Encounter: 12601 to 13000
    #Zen Encounter: 13101 to 13100

```
