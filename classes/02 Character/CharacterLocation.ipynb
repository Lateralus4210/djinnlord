```python
class CharacterLocation:
    
    position_current = [0,0]
    distance_traveled = 0
    map_explored = len(Map.map_explored_dict)/2601
    # max x and max y are 25 and -25
    # (51)^2 = 2601

    ####################################################################################
            
    def increment_character_position_current_x():
        CharacterLocation.position_current[0] += 1
        CharacterLocation.refresh()

    def increment_character_position_current_y():
        CharacterLocation.position_current[1] += 1
        CharacterLocation.refresh()

    def decrement_character_position_current_y():
        CharacterLocation.position_current[1] -= 1
        CharacterLocation.refresh()

    def decrement_character_position_current_x():
        CharacterLocation.position_current[0] -= 1
        CharacterLocation.refresh()

    def refresh():
        InterfaceUpdate.update_character_position_current_label()
        Map.check_map_explored_dict()
        TerrainTables.check()

    ####################################################################################    

    def increment_character_position_current_x_twice():
        CharacterLocation.position_current[0] += 1
        Map.check_map_explored_dict()
        CharacterLocation.position_current[0] += 1
        Map.check_map_explored_dict()

    def decrement_character_position_current_x_twice():
        CharacterLocation.position_current[0] -= 1
        Map.check_map_explored_dict()
        CharacterLocation.position_current[0] -= 1
        Map.check_map_explored_dict()

    def increment_character_position_current_y_twice():
        CharacterLocation.position_current[1] += 1
        Map.check_map_explored_dict()
        CharacterLocation.position_current[1] += 1
        Map.check_map_explored_dict()

    def decrement_character_position_current_y_twice():
        CharacterLocation.position_current[1] -= 1
        Map.check_map_explored_dict()
        CharacterLocation.position_current[1] -= 1
        Map.check_map_explored_dict()

    ####################################################################################    

    def update_character_position_current(direction):
        
        if direction in ValidInputs.north_inputs:
            if CharacterLocation.position_current[1] < 25: 
                CharacterLocation.increment_character_position_current_y()
            if CharacterLocation.position_current[1] == 25:
                CommandsFrame.entry_listbox.insert(0, "You can't go any farther.")
            
        if direction in ValidInputs.east_inputs:
            if CharacterLocation.position_current[0] < 25: 
                CharacterLocation.increment_character_position_current_x()
            if CharacterLocation.position_current[0] == 25:
                CommandsFrame.entry_listbox.insert(0, "You can't go any farther.")
            
        if direction in ValidInputs.south_inputs:
            if CharacterLocation.position_current[1] > -25: 
                CharacterLocation.decrement_character_position_current_y()
            if CharacterLocation.position_current[1] == -25:
                CommandsFrame.entry_listbox.insert(0, "You can't go any farther.")
            
        if direction in ValidInputs.west_inputs:
            if CharacterLocation.position_current[0] > -25: 
                CharacterLocation.decrement_character_position_current_x()
            if CharacterLocation.position_current[0] == -25:
                CommandsFrame.entry_listbox.insert(0, "You can't go any farther.")

    #######################################################################################
        
        if direction in ValidInputs.dash_north_inputs:
            if CharacterLocation.position_current[1] < 24: 
                CharacterLocation.increment_character_position_current_y_twice()
            if CharacterLocation.position_current[1] == 24:
                CommandsFrame.entry_listbox.insert(0, "You can't dash that far.")
            
        if direction in ValidInputs.dash_east_inputs:
            if CharacterLocation.position_current[0] < 24: 
                CharacterLocation.increment_character_position_current_x_twice()
            if CharacterLocation.position_current[0] == 24:
                CommandsFrame.entry_listbox.insert(0, "You can't dash that far.")
            
        if direction in ValidInputs.dash_south_inputs:
            if CharacterLocation.position_current[1] > -24: 
                CharacterLocation.decrement_character_position_current_y_twice()
            if CharacterLocation.position_current[1] == -24:
                CommandsFrame.entry_listbox.insert(0, "You can't dash that far.")
            
        if direction in ValidInputs.dash_west_inputs:
            if CharacterLocation.position_current[0] > -24: 
                CharacterLocation.decrement_character_position_current_x_twice()
            if CharacterLocation.position_current[0] == -24:
                CommandsFrame.entry_listbox.insert(0, "You can't dash that far.")
        
        InterfaceUpdate.update_character_position_current_label()
        EventDetermine.random()

        CharacterLocation.distance_traveled += 1
        StatisticsFrame.distance_traveled_label.config(text="Distance traveled: " + str(CharacterLocation.distance_traveled))

        #update_visible_map()
```
