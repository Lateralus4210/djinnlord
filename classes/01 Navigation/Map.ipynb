```python
import csv

class Map:
    
    # Reads the map data CSV, then stores it in a dict for reference later. [csv_file_path and map_dict]
    
    # Expects a CSV file delimited by semicolons (;) that contains -25 to 25 in the first row and -25 to 25 in the first column.
    # Cell A1 (row 1, column 1) is empty.

    # 
    
    csv_file_path = 'W:/ELECTRONICS & CS/CODE - PROGRAMMING/Python/Projects/00 Djinnlord/Map data/Djinnlord Map prototype 2.csv'
    map_dict = {}
    map_explored_dict = {}
    
    @classmethod
    def read_csv_to_dict(cls, csv_file_path):
        with open(csv_file_path, newline='') as csvfile:
            reader = csv.reader(csvfile, delimiter=';')
            next(reader)  # Skip the first row (X-coordinates labels)
            
            for x, row in enumerate(reader, start=-25):
                cls.map_dict[x] = {}
                for y, value in enumerate(row[1:], start=-25):  # Skip the first column (Y-coordinate label)
                    if value:
                        params = [param.strip() for param in value.split(',')]
                        if len(params) >= 1:
                            map_object = {
                                "terrain": params[0]
                            }
                            cls.map_dict[x][y] = map_object
                        else:
                            print(f"Not enough parameters in item '{item}' at ({x}, {y})")
                            cls.map_dict[x][y] = None
                    else:
                        print(f"Empty item at ({x}, {y})")
                        cls.map_dict[x][y] = None

    @classmethod
    def get_object_at_coordinate(cls, x, y):
        debug_listbox.insert(0, cls.map_dict.get(y, {}).get(x, None), (str(x), str(y)))
        debug_listbox.insert(0, "Map object called: ")
        debug_listbox.insert(0, "-------------------")
        return cls.map_dict.get(y, {}).get(x, None)

    def check_map_explored_dict():
        position_to_check = tuple(CharacterLocation.position_current)
        already_explored = False
    
        for position in Map.map_explored_dict:
            if position == position_to_check:
                already_explored = True
                CommandsFrame.entry_listbox.insert(0, "This place is familiar.")
                debug_listbox.insert(0, "Already explored.")
                break
    
        if not already_explored and position_to_check != (0,0):
            Map.map_explored_dict[position_to_check] = None
            CommandsFrame.entry_listbox.insert(0, "You haven't been here before. " + str(position_to_check))
            debug_listbox.insert(0, "New coordinate added.")

Map.read_csv_to_dict(Map.csv_file_path)

# from pprint import pprint
pprint(Map.map_dict)
```
