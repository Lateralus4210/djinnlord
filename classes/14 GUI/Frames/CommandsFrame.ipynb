```python
class CommandsFrame:
    screen_height = root.winfo_screenheight()
    screen_width = root.winfo_screenwidth()
    one_third_screen_width = screen_width // 3
    one_third_screen_height = screen_height // 3
    commands_frame         = tk.Frame(root, borderwidth=0, relief="flat", padx=0, pady=0, bg=Palettes.panel_one_color)

    
    commands_frame.place (x= one_third_screen_width,
              y= one_third_screen_height,
              width = one_third_screen_width - InterfaceStyleConfig.margin_pixel_amount,
              height = 2*one_third_screen_height - InterfaceStyleConfig.margin_pixel_amount)
              

    

    commands_label = tk.Label(commands_frame, text="Commands")
    commands_label.pack()

    InterfaceStyleConfig.title_labels.append(commands_label)
    InterfaceStyleConfig.panel_three_labels.append(commands_label)

    ################################################################
    
    entry_text = tk.StringVar()
    entry = tk.Entry(commands_frame, 
                     textvariable=entry_text, 
                     width=30,
                     font=Fonts.main_title_font,
                     insertbackground=Palettes.panel_three_text_color)
    InterfaceStyleConfig.panel_three_labels.append(entry)
    
    def convert_to_uppercase(*args):
        CommandsFrame.entry_text.set(CommandsFrame.entry_text.get().upper())
    
    entry_text.trace_add("write", convert_to_uppercase)
    entry.pack()
    
    entry.bind("<Return>", Interpret.handle_user_input)
    
    #submit_button = tk.Button(commands_frame, text="Enter", bg=panel_three_color, fg=panel_one_text_color,  activebackground="#111111",  # Set the background color for the active state
    #    activeforeground="#FFFFFF",  command=handle_user_input)
    #submit_button.pack(padx=10, pady=10)
    
    entry_listbox_frame = tk.Frame(commands_frame, bg=Palettes.panel_three_color, padx=20, pady=20)
    entry_listbox_frame.pack(fill=tk.BOTH, expand=True)
    
    entry_listbox = tk.Listbox(entry_listbox_frame,
                               font=Fonts.sub_title_font,
                               bd=1,
                               selectmode=tk.NONE)
                               #yscrollcommand=yscroll.set)
    InterfaceStyleConfig.panel_three_labels.append(entry_listbox)
    entry_listbox.pack(fill=tk.BOTH, expand=True)
    
    def on_item_select(event):
        return "break"  # Prevents item selection
    
    entry_listbox.bind("<Button-1>", on_item_select)
    
    #yscroll = Scrollbar(entry_listbox, orient=tk.VERTICAL)
    #yscroll.config(command=entry_listbox.yview)
    #yscroll.pack(side="right", anchor="e", fill=tk.Y, expand=True)
    # yscroll.set(0, 1)
    
    
    
```
