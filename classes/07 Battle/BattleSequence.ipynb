```python
class BattleSequence:
    
    def character_attack_attempt(spell_name):
    
        CharacterSkills.determine_character_attack_accuracy()
        
        random_player_accuracy_number = random.randint(1, 100)
        #print("Player accuracy: " + str(player_attack_accuracy))
    
        if random_player_accuracy_number <= CharacterSkills.attack_accuracy:

            # time.sleep(Interpret.pause_amount)
            # CommandsFrame.entry_listbox.insert(0, "Casting...")
              
            time.sleep(Interpret.pause_amount)
            CommandsFrame.entry_listbox.insert(0, "Successful cast!")


            Spell.cast_spell(spell_name)
            # update_enemy_health_progressbar()

            if Enemy.health_current > 0:
                BattleSequence.enemy_attack_attempt()

            if Enemy.health_current <= 0:
                Enemy.health_current = 0
                CharacterProfile.in_battle = False
                BattleSuccess.go()
                
    
        else:
            # print("Player missed the attack.")

            time.sleep(Interpret.pause_amount)
            CommandsFrame.entry_listbox.insert(0, "Casting...")
            time.sleep(Interpret.pause_amount)
            CommandsFrame.entry_listbox.insert(0, "The attack missed this time.")
            time.sleep(Interpret.pause_amount)
            
            BattleSequence.enemy_attack_attempt()
        

    def enemy_attack_attempt():
        
        random_enemy_accuracy_number = random.randint(1, 100)

        if CharacterSkills.level_current < 40:
            Enemy.accuracy = 50

        elif CharacterSkills.level_current >= 40:
            Enemy.accuracy = 70

        elif CharacterSkills.level_current >= 65:
            Enemy.accuracy = 85
    
        if random_enemy_accuracy_number <= Enemy.accuracy:

            time.sleep(Interpret.pause_amount)
            CommandsFrame.entry_listbox.insert(0, "The enemy attacks...")

            time.sleep(Interpret.pause_amount)
            CommandsFrame.entry_listbox.insert(0, "The enemy hit you!")
            EnemyAttack.go()
    
        if random_enemy_accuracy_number > Enemy.accuracy:
            # print("Player missed the attack.")

            time.sleep(Interpret.pause_amount)
            CommandsFrame.entry_listbox.insert(0, "The enemy attacks...")
            time.sleep(Interpret.pause_amount)
            CommandsFrame.entry_listbox.insert(0, "The enemy missed.")
        
```
