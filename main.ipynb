#### Variable Lists

> Grouped by class.

```python
# Get the global variables in the current session
global_variables = list(globals().items())

# Sort the list of global variables by their "Type" column
sorted_result = sorted(global_variables, key=lambda x: str(type(x[1])))

# Print the sorted result
for name, variable in sorted_result:
    print(f"{name:30} {str(type(variable)):20}")
```

> Grouped alphabetically.

```python
# Run the %whos magic function and store its output in a variable
whos_result = %whos

# Check if whos_result is not None
if whos_result is not None:
    # Parse and extract the relevant information
    variable_info = []
    for entry in whos_result:
        variable_name = entry[0]
        variable_type = entry[2]
        variable_size = entry[3]
        
        # Check if the variable type is "NoneType" (None)
        if variable_type == 'NoneType':
            continue  # Skip NoneType variables
        
        # Remove the "whos_result" variable from the list
        if variable_name == 'whos_result':
            continue
        variable_info.append((variable_name, variable_type, variable_size))

    # Sort the list of variable information by the "Data/Info" column (size)
    sorted_result = sorted(variable_info, key=lambda x: x[2], reverse=False)

    # Print the sorted result
    for item in sorted_result:
        print(f"Name: {item[0]}, Type: {item[1]}, Size: {item[2]}")
```
